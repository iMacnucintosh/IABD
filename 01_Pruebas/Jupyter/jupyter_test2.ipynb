```python
# ¡Bienvenido a tu primer cuaderno Jupyter de Python!

# En este cuaderno, aprenderás conceptos básicos como variables, operaciones, listas, funciones, condicionales, bucles y visualización de datos.

# 1. Variables y Tipos de Datos

# Asignamos diferentes tipos de datos a variables
x = 10          # Entero
y = 3.14        # Flotante
nombre = "Juan" # Cadena de texto

# Mostramos los valores de las variables
x, y, nombre

# 2. Operaciones Matemáticas

# Realizamos algunas operaciones matemáticas básicas
suma = x + y
resta = x - y
multiplicacion = x * y
division = x / y

# Mostramos los resultados
suma, resta, multiplicacion, division

# 3. Listas

# Creamos una lista de frutas
frutas = ["manzana", "banana", "cereza"]

# Accedemos al primer elemento de la lista
primer_fruta = frutas[0]

# Añadimos una nueva fruta a la lista
frutas.append("naranja")

# Mostramos la lista actualizada
frutas

# 4. Funciones

# Definimos una función que saluda a una persona
def saludar(nombre):
    return f"Hola, {nombre}!"

# Llamamos a la función pasando un argumento
saludar(nombre)

# 5. Condicionales

# Usamos una sentencia condicional para comparar dos valores
if x > 5:
    resultado = "x es mayor que 5"
else:
    resultado = "x es menor o igual a 5"

resultado

# 6. Bucles

# Usamos un bucle for para iterar sobre una lista de frutas
for fruta in frutas:
    print(fruta)

# 7. Visualización de Datos con Matplotlib

# Importamos la librería para crear gráficos
import matplotlib.pyplot as plt

# Datos para un gráfico de ejemplo
x_vals = [1, 2, 3, 4, 5]
y_vals = [1, 4, 9, 16, 25]

# Creamos un gráfico de líneas
plt.plot(x_vals, y_vals)
plt.title("Gráfico de ejemplo")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.grid(True)
plt.show()

# 8. Uso de Pandas para Manejo de Datos

# Importamos pandas para trabajar con tablas de datos
import pandas as pd

# Creamos un DataFrame con datos de ejemplo
data = {"Nombre": ["Ana", "Luis", "Carlos"], "Edad": [23, 34, 45]}
df = pd.DataFrame(data)

# Mostramos el DataFrame
df
```
